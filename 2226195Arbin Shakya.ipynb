Write code to transform variables according to the following instructions:

1. ”School”, ”sex”, ”address”, ”schoolsup”, ”famsup”, ”activites”, ”nursery”, ”internet”, and ”romantic”
into binary: 0 or 1 (create new columns without overwriting the existing ones).

```python
columns = ["school","sex","address","schoolsup","famsup","activities","nursery","internet","romantic"]

for col in columns:
    unique_values = df[col].unique()
    
    for value in unique_values:
        new_col_name = f"{col}_{value}"
        df[new_col_name] = (df[col] == value).astype(int)
df
```

2. ”Medu”, ”Fedu”, ”reason”, ”guardian”, ”studytime”, ”freetime”, and ”health” into ordinal numbers based
on the number cases in the data set (create news columns without overwriting the existing ones).

```python
column = ["Medu", "Fedu", "reason", "guardian", "studytime", "freetime", "health"] # making array

for c in column: #looping in column array
  values = df[c].value_counts() # searches for the unique values
    
  valuemap = {value: i+1 for i, value in enumerate(values.index)} 
    
  newcolumns = f"{c}_ordinal" # creating new columns 
  df[newcolumns] = df[c].map(valuemap) # maping the ordinal values 
df # displays df
```

3. Convert column ”age” to interval datatype. i.e. Create a new column name category age whose values
should be based on the frequency in the column ”age”, You can create categorical data with following
interval.
(a) interval1: [15-17]
(b) interval2: [18-20]
(c) interval3: [21-all]

```python
df['category_age'] = "" # creating a new column 
df.category_age = pd.IntervalDtype(subtype='int64') # changing into interval datatype


df.loc[(df['age'] <= 17) & (df['age'] >=15), 'category_age'] = "[15-17]" # changing the value to interval 15-17  
df.loc[(df['age'] > 17) & (df['age'] <= 20), 'category_age'] = "[17-20]" # changing the value to interval 17-20
df.loc[df['age'] > 20, 'category_age'] = '[20-all]' # changing the value to interval 20-all 

df # displaying new df
```

4. Create a new column name passed (yes or no) whose values should be based on the values present in the
G3 column (>= 8–yes, < −no).

```python
df["passed"] = "" # creating a new column named as passed


df.loc[df["G3"] >= 8, "passed"] = "yes" # storing yes in passed column if the value of G3 is greater than or equal to 8
df.loc[df["G3"] < 8, "passed"] = "no" # storing no in passed column if the value of G3 is less than 8


df
```
